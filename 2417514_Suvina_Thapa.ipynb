Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
Q1_asia_score = south_country_df['score'].quantile(0.25)
Q3_asia_score = south_country_df['score'].quantile(0.75)
IQR_asia_score = Q3_asia_score - Q1_asia_score

#to findquartile for gdp
Q1_asia_gdp = south_country_df['Log GDP per capita'].quantile(0.25)
Q3_asia_gdp = south_country_df['Log GDP per capita'].quantile(0.75)
IQR_asia_gdp = Q3_asia_gdp - Q1_asia_gdp

#For upper and lower bound for outlier to lie
upper_score = Q3_asia_score + 1.5 * IQR_asia_score
lower_score = Q1_asia_score - 1.5 * IQR_asia_score

upper_gdp = Q3_asia_gdp + 1.5 * IQR_asia_gdp
lower_gdp = Q1_asia_gdp - 1.5 * IQR_asia_gdp

#Identifying outlier
outlier_identify = south_country_df[
   ((south_country_df['score'] > upper_score) | (south_country_df['score'] < lower_score))  &
   ((south_country_df['Log GDP per capita'] > upper_gdp) | (south_country_df['Log GDP per capita'] < lower_gdp)
    )
]
outlier_identify[['Country name','score','Log GDP per capita']]
```

6. Outlier Detection:
â€¢ Identify outlier countries in both regions based on Score and GDP per Capita.

```python
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

south_asia_score_outliers = find_outliers_iqr(south_country_df, 'score')
south_asia_gdp_outliers = find_outliers_iqr(south_country_df, 'Log GDP per capita')


middle_east_score_outliers = find_outliers_iqr(middle_eastern_df, 'score')
middle_east_gdp_outliers = find_outliers_iqr(middle_eastern_df, 'Log GDP per capita')

print("Discussion :")
print("\nSouth Asia Outliers:")
if len(south_asia_score_outliers) > 0:
    for index, row in south_asia_score_outliers.iterrows():
        print(f"Outlier Country: {row['Country name']}, Score: {row['score']}, GDP per Capita: {row['Log GDP per capita']}")
else:
    print("No outliers found based on Score in South Asia.")
print('')
print("Middle East Outliers:")
if len(middle_east_score_outliers) > 0:
    for index, row in middle_east_score_outliers.iterrows():
        print(f"Outlier Country: {row['Country name']}, Score: {row['score']}, GDP per Capita: {row['Log GDP per capita']}")
else:
    print("No outliers found based on Score in Middle East.")
```


plot outlier
